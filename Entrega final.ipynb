```python
#Importaremos la tabla
import pandas as pd
import os

datadis=pd.read_excel(os.path.join('data','dataPerú.xlsx'),
                     dtype={'Ubigeo':object})
datadis.head()
```

```python
# el mapa es el siguiente
os.environ['USE_PYGEOS'] = '0'
import geopandas as gpd
datadismap=gpd.read_file(os.path.join('maps','DistritosMap'))
datadismap.head()
```

```python
#Realizamos, si es necesario, una limpieza y un formateo
#Verificamos que todos los nombres estén escritos igualmente 
capitalizeColumns=lambda x: x.str.upper().str.strip()
#Convertimos todos los nombres en mayúsculas
datadis[['Provincia','Distrito']]=datadis[['Provincia','Distrito']].apply(capitalizeColumns)
datadismap[['PROVINCIA','DISTRITO']]=datadismap[['PROVINCIA','DISTRITO']].apply(capitalizeColumns)
```

```python
import unidecode
#Para eliminar símbolos especiales
byePunctuation=lambda x: unidecode.unidecode(x)
datadis[['Provincia','Distrito']]=datadis[['Provincia','Distrito']].applymap(byePunctuation)
datadismap[['PROVINCIA','DISTRITO']]=datadismap[['PROVINCIA','DISTRITO']].applymap(byePunctuation)
```

```python
len(datadis.Distrito),len(datadismap.DISTRITO)
```

```python
#Debemos verificar que no haya repetidos porque si es así, no funcionará
datadis.Distrito.duplicated().sum(),datadismap.DISTRITO.duplicated().sum()
```

```python
#Añadimos columnas para diferenciar distritos, ya que algunos son repetidos
#Concatenamos
datadis['provDist']=["+".join(pd) for pd in zip (datadis.Provincia,datadis.Distrito)]
datadismap['provDist']=["+".join(pd) for pd in zip (datadismap.PROVINCIA,datadismap.DISTRITO)]
```

```python
#Cambiando los slashes, guiones arriba y abajo por espacios simples
datadis.provDist=datadis.provDist.str.replace("\-|\_|\s+"," ",regex=True)
datadismap.provDist=datadismap.provDist.str.replace("\-|\_|\s+"," ",regex=True)
```

```python
#Así se vería la nueva columna
datadis['provDist'].head()
```

```python
#Ahora vemos si hay filas mal escritas
nomatch_df=set(datadis.provDist)- set(datadismap.provDist)
nomatch_gdf=set(datadismap.provDist)-set(datadis.provDist)
```

```python
len(nomatch_df), len(nomatch_gdf)
```

```python
#Cuáles se asemejan más (identificar difusos)
from thefuzz import process
[(dis,process.extractOne(dis,nomatch_gdf)) for dis in sorted(nomatch_df)]
```

```python
#Cambiaremos algunos nombres
{dis:process.extractOne(dis,nomatch_gdf)[0] for dis in sorted(nomatch_df)}
```

```python
#Los cambiamos
changesDis_df={dis:process.extractOne(dis,nomatch_gdf)[0] for dis in sorted(nomatch_df)}
```

```python
#Reemplazamos
datadis.provDist.replace(changesDis_df,inplace=True)
```

```python
#Merging
#Añadir la tabla a la mapa (df->gdf)
datadismap=datadismap.merge(datadis, on='provDist')
```

```python
datadismap.info()
```

```python
#Borramos las columnas que no necesitamos y las repetidas
bye=['Departamento', 'Provincia', 'Distrito','INSTITUCIO']
datadismap.drop(columns=bye,inplace=True)
#Verificamos
datadismap.head()
```

```python
import mapclassify 
import numpy as np

np.random.seed(12345) #utilizamos el siguiente número

#usaremos 5 intervalos
K=5
#Mismos intervalos pero con una interpretación sencilla
ei5 = mapclassify.EqualInterval(lima['IPT'], k=K)
# Basado en la desviación estándar
msd = mapclassify.StdMean(lima['IPT'])
# Ancho del intervalo varía                              
q5=mapclassify.Quantiles(lima['IPT'],k=K)

#Similitud, basado en datos multimodales
mb5 = mapclassify.MaximumBreaks(lima['IPT'], k=K)
#Para datos sesgados
ht = mapclassify.HeadTailBreaks(lima['IPT']) # no K needed
#Basados en similitud, optimizador
fj5 = mapclassify.FisherJenks(lima['IPT'], k=K)
jc5 = mapclassify.JenksCaspall(lima['IPT'], k=K)
mp5 = mapclassify.MaxP(lima['IPT'], k=K)   
```

```python
#Se calcula con el vecindario
#Si el valor de un distrito se encuentra correlacionado con un distrito
#Variable afectada por su vecindario
#La correlación espacial es medida por el Moran
#Morán para todas mis variables numéricas
from esda.moran import Moran

moranIPT = Moran(lima['IPT'], w_knn)
moranIPT.I,moranIPT.p_sim
```
